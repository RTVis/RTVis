```python
import pandas as pd
from collections import Counter
```

```python
# read stopwords
sw = pd.read_csv('stop_words.csv')
sw = sw['0'].tolist()
sw.append('0')
        
# drop the float (nan)
sw = [i for i in sw if type(i) != float]

# sort in the order of letter
sw = sorted(sw)

# count duplicates
c = Counter(sw)

# drop duplicates
sw = list(dict.fromkeys(sw))

print(c)
print(sw)
```

```python
newsw = []
# keep only letters and numbers
for i in sw:
    for j in i:
        if not (j.isalpha() or j.isdigit()):
            print(i)
            i = i.replace(j, '')
    newsw.append(i)

print(newsw)
```

```python
# save to csv
pd.DataFrame(newsw).to_csv('stop_words.csv', index=False, header=False)
```
